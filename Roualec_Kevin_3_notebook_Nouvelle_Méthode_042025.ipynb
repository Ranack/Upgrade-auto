# Cityscapes_1 (C1)

```python
import requests
import os

# URL du fichier à télécharger
url = 'https://zenodo.org/records/14766160/files/vltseg_checkpoint_cityscapes_1.pth?download=1'

# Chemin de destination sur Google Colab
destination_path = 'checkpoints/vltseg_checkpoint_cityscapes_1.pth'

# Créer le répertoire de destination s'il n'existe pas
os.makedirs(os.path.dirname(destination_path), exist_ok=True)

# Télécharger le fichier
response = requests.get(url)

# Vérifier si la requête a réussi
if response.status_code == 200:
    with open(destination_path, 'wb') as file:
        file.write(response.content)
    print("Fichier téléchargé et enregistré avec succès !")
    print(f"Chemin du fichier : {destination_path}")
else:
    print(f"Échec du téléchargement. Code d'état : {response.status_code}")
```

```python
from google.colab import files
import shutil
import os

# Vérifier si le fichier existe et le supprimer s'il est présent
if os.path.exists('demo_composition3.py'):
    os.remove('demo_composition3.py')
    print("Le fichier demo_composition3.py a été supprimé.")

# Télécharger le fichier depuis l'ordinateur local
uploaded = files.upload()

# Vérifier si le fichier a été téléchargé
if 'demo_composition3.py' in uploaded:
    # Déplacer le fichier téléchargé vers le dossier VLTSEG
    shutil.move('demo_composition3.py', '/content/VLTSeg/demo_composition3.py')
    print("Fichier remplacé avec succès !")

    # Afficher le chemin du fichier déplacé
    file_path = os.path.abspath('/content/VLTSeg/demo_composition3.py')
    print(f"Le fichier a été déplacé vers : {file_path}")
else:
    print("Aucun fichier téléchargé ou nom de fichier incorrect.")
```
